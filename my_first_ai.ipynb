```python
import pandas as pd
import numpy as np
```

```python
examples = 20000
X = [] 
for i in range(examples):
    X.append(np.random.randint(2, size=9))
X
```

```python
Y = []
for i in range(examples):
    if  X[i][0] == X[i][8]:
        Y.append([1, 0])
    elif X[i][1] == X[i][3] and X[i][3] == X[i][6]:
        Y.append([1,1])
    else:
        Y.append([0,1])
Y
```

```python
import tensorflow as tf
import tflearn
```

```python
def build_model():
    tf.reset_default_graph()
    
    #inputs
    net = tflearn.input_data([None, 9])
    
    #hidden layer
    net = tflearn.fully_connected(net, 40, activation='ReLU')
    net = tflearn.fully_connected(net, 40, activation='ReLU')
    net = tflearn.fully_connected(net, 20, activation='ReLU')
    
    #output layer
    net = tflearn.fully_connected(net, 2, activation='softmax')
    net = tflearn.regression(net, optimizer='sgd', learning_rate=0.05, loss='categorical_crossentropy')
    model = tflearn.DNN(net)
    return model
```

```python
def translate(array):
    if array[0][0] > array[0][1]:
        return [1,0]
    return [0,1]
```

```python
model = build_model()
```

```python
# Training
model.fit(X, Y, validation_set=0.1, show_metric=True, batch_size=128, n_epoch=40)
```

```python
model.predict([[0,1,0,0,1,0,1,1,1]])
```
